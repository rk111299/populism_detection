```python
# Function from google sheet to pre-processed pandas dataframe - work out how to get around reason column to int!
def sheet_preprocessing(worksheet):
  rows = worksheet.get_all_values()
  df = pd.DataFrame(rows)
  df.columns = df.iloc[0]
  df = df.iloc[1:]
  df.loc[ df['populist'] == '1 - Populist', 'populist'] = 1
  df.loc[ df['populist'] == '0 - Not Populist', 'populist'] = 0
  df.loc[df['reason'].str.contains('1', na=False), 'reason'] = 1
  df.loc[df['reason'].str.contains('2', na=False), 'reason'] = 2
  df.loc[df['reason'].str.contains('3', na=False), 'reason'] = 3
  df.loc[df['reason'].str.contains('4', na=False), 'reason'] = 4
  df.loc[df['reason'].str.contains('5', na=False), 'reason'] = 5
  df.loc[df['reason'].str.contains(' ', na=False), 'reason'] = 0
  df['populist'] = df['populist'].astype(int)
  #df['reason'] = df['reason'].astype(int)
  return df
```

```python
# How to do cohen kappa score
from sklearn.metrics import cohen_kappa_score

def cohen_kappa(df1, df2):
  anno1 = df1['populist'].to_numpy()
  anno2 = df2['populist'].to_numpy()
  score = cohen_kappa_score(anno1, anno2)
  return score
```

```python
def fleiss_kappa(M):
    """Computes Fleiss' kappa for group of annotators.
    :param M: a matrix of shape (:attr:'N', :attr:'k') with 'N' = number of subjects and 'k' = the number of categories.
        'M[i, j]' represent the number of raters who assigned the 'i'th subject to the 'j'th category.
    :type: numpy matrix
    :rtype: float
    :return: Fleiss' kappa score
    """
    N, k = M.shape  # N is # of items, k is # of categories
    n_annotators = float(np.sum(M[0, :]))  # # of annotators
    tot_annotations = N * n_annotators  # the total # of annotations
    category_sum = np.sum(M, axis=0)  # the sum of each category over all items

    # chance agreement
    p = category_sum / tot_annotations  # the distribution of each category over all annotations
    PbarE = np.sum(p * p)  # average chance agreement over all categories

    # observed agreement
    P = (np.sum(M * M, axis=1) - n_annotators) / (n_annotators * (n_annotators - 1))
    Pbar = np.sum(P) / N  # add all observed agreement chances per item and divide by amount of items

    return round((Pbar - PbarE) / (1 - PbarE), 4)
```

Combining all annotated tweets into one dataframe and writing to csv

```python
# Importing theo's original tweets and labels 
pop_labelled = pd.read_csv('/content/drive/MyDrive/Diss_data/annotated_populist_tweets.csv')[['text', 'label']]

pop_labelled.rename(columns = {'label':'populist'}, inplace = True)
pop_labelled
```

```python
batchnames = ['rob_2017batch', 'rob_2019batch', 'rob_batch1',
              'Alex_Annotations', 'Ben_Annotations', 'Chloe_Annotations',
              'Elli_Annotations', 'Emily_Annotations', 'Izzy_Annotations',
              'Jed_Annotations', 'Josh_Annotations', 'Laura_Annotations',
              'Nancy_B_Annotations', 'Nancy_LV_Annotations', 'Zella_Annotations']
```

```python
def importing_annotations(batchnames):
  
  # Empty dataframe
  all_anno_df = pd.DataFrame()

  # for loop to go through google sheets and combine populist labels
  for i in batchnames:

    # open google sheet  
    batch = gc.open(i).sheet1
    # pre process sheet with previously defined function
    df = sheet_preprocessing(batch)
    # select only text and populist columns
    df = df[['text', 'populist']]

    # append these to new empty dataframe
    all_anno_df = all_anno_df.append(df, ignore_index = True)

  return all_anno_df
```

```python
# Compiling final populist df
final_populist_tweets = importing_annotations(batchnames)
final_populist_tweets = final_populist_tweets.append(pop_labelled, ignore_index=True)
final_populist_tweets.rename(columns = {'populist':'label'}, inplace = True)
final_populist_tweets
```
